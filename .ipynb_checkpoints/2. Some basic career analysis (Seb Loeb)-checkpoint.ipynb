# Find smaller accidents and come-backs

```python
grouped = df.groupby(['Name','Year'])
results = []

for name, group in grouped:
    group['diff'] = group['stage'].diff()
    
    for index, row in group.iterrows():
        if row['diff'] > 20 or row['diff'] < -20:
            results.append([row['Name'], row['Year'], row['ssname'], row['diff']])
            
results
```
