```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from urllib.parse import urlencode
import urllib.parse
from copy import copy
import requests
import re
import json
import urllib.request
```

```python
df1 = pd.read_csv('Sébastien_Loeb.csv') 
df2 = pd.read_csv('Sébastien_Ogier.csv') 
df3 = pd.read_csv('./Other_Drivers/Carlos_Sainz.csv') 
df4 = pd.read_csv('./Other_Drivers/Colin_McRae.csv') 
df5 = pd.read_csv('./Other_Drivers/Elfyn_Evans.csv') 
df6 = pd.read_csv('./Other_Drivers/Mikko_Hirvonen.csv') 
df7 = pd.read_csv('./Other_Drivers/Ott_Tänak.csv') 
df8 = pd.read_csv('./Other_Drivers/Thierry_Neuville.csv') 
df9 = pd.read_csv('./Other_Drivers/Tommi_Mäkinen.csv') 

master_df = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8, df9])
```

```python
master_df.head(5)
```

```python
master_df = master_df.drop(['Unnamed: 0', 'Php_url','p_overall','stage','ssname'], axis=1)
master_df.head(5)
```

```python
master_df = master_df.drop_duplicates()
master_df.head(5)
```

```python
def extract_race_name(url):
    match = re.search(r"/entryinfo/(.+)/.+/", url)
    if match:
        return match.group(1)
    else:
        return "Unknown"
```

```python
master_df["Match"] = master_df["URL"].apply(extract_race_name)
master_df
```

# Call the race df

```python
race_df = pd.read_csv('race_list.csv') 
race_df
```

# Merge two dfs

```python
merged_df = master_df.merge(race_df, left_on='Match', right_on='Name', how='left')
merged_df = merged_df.sort_values(by='Serial')
merged_df
```

```python
merged_df.to_csv('Merged.csv')
```
